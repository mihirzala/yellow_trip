```python
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import numpy as np
# import plotly.express as px
```

```python
date_pickup = df['tpep_pickup_datetime'].nunique()
print('date_pickup',date_pickup)

date_dropoff = df['tpep_dropoff_datetime'].nunique()
print('date_dropoff',date_dropoff)
```

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
# pickup from day, month, date, hour, min, weekday
df['pickup_day'] = df['tpep_pickup_datetime'].dt.day
df['pickup_month'] = df['tpep_pickup_datetime'].dt.month
df['pickup_date'] = df['tpep_pickup_datetime'].dt.date
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
df['pickup_min'] = df['tpep_pickup_datetime'].dt.minute
df['pickup_weekday'] = df['tpep_pickup_datetime'].dt.weekday
df['dropoff_weekday'] = df['tpep_dropoff_datetime'].dt.weekday

#creating new column of drop off in minute as it will easy to read the data.
df['dropoff_min'] = df['tpep_dropoff_datetime'].dt.minute
```

```python
# creating target variable named as trip_duration
df['trip_duration'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
df['trip_duration'] = df['trip_duration'] / np.timedelta64(1, 'm')
```
